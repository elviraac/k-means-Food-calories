```python
#Import Library yang akan digunakan
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#Menyiapkan data dan memanggil dataset
dataset = pd.read_csv('FoodCalories.csv')
dataset.keys()
```

```python

dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
#konversi ke data Array
X = np.asarray(dataset)
print(X)
```

```python
#Menampilkan data dlam bentuk scatter plot
plt.scatter(X[:,0], X[:,1], label='True Position')
plt.xlabel("serving")
plt.ylabel("calories")
plt.title("Grafik kalori makanan perserving")
plt.show()
```

```python
 #Mengaktifkan K-Means dengan jumlah K=2
 kmeans = KMeans(n_clusters=2)
 kmeans.fit(X)
```

```python
#Menampilkan nilai centroid yang digenerate oleh algoritma 
print(kmeans.cluster_centers_)
```

```python
#Plot Data Point 
#Memvisualisasikan Hasil Klatering Data Konsumen
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("serving")
plt.ylabel("calories")
plt.title("Grafik kalori makanan perserving")
plt.show()
```

```python
#PLot Data Point
#Memvisualisasikan hasil klastering dengan centroid dr masing2 klaster
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("serving")
plt.ylabel("calories")
plt.title("Grafik kalori makanan perserving")
plt.show()
```
